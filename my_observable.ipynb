```
from jupyter_dash import JupyterDash

import dash
from dash.dependencies import Input, Output
import dash_core_components as dcc
import dash_html_components as html

import my_observable
```

```
app = JupyterDash(__name__)

# Configure the dash slider
dash_slider = dcc.Slider(
    id='dash-slider',
    min=0,
    max=2,
    step=.1,
    value=.1,  # callback uses this as the initial speed of the animation
    updatemode='drag',
)

# Create the app layout
app.layout = html.Div([
    my_observable.MyObservable(
        id='my-observable',
        speed=0 # initial value (will be overriden by callback)
    ),
    dash_slider,
])

# When the slider value changes, use it to update the animation speed
@app.callback(Output('my-observable', 'speed'), [Input('dash-slider', 'value')])
def send_speed(value):
    return value
```

```
# Run the app inside a tab in this notebook
app.run_server(mode="inline")
```
